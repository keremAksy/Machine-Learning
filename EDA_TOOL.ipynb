## Examining the data distribution, data anomalies and correlations. A more detailed exploration

```python
# Plotting each of the variables
for i in range(D):
  sns.displot(df, x=df.columns[i])
```

```python
# Plotting histogram for a variable
df['longitude'].plot(kind='hist', bins=50, figsize=(12,6), facecolor='grey',edgecolor='black')
```

Checking Correlation Between Variables(features)

```python
corrMatrix = df.corr() #This returns DxD correlation matrix for the variables
sns.heatmap(corrMatrix, annot = True, cmap= 'coolwarm') #Visualizing it by using a heat map
```

```python
#Scatter plot of the pairwise correlation of the variables
# An example
variable1 ='total_rooms' 
variable2 ='median_house_value'
df.plot(kind='scatter', x=variable1, y=variable2)
```

```python
#This function plots(scatters) all the pairwise correlation between variables 
sns.pairplot(df)
```
